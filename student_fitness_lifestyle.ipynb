```python
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# Load Excel data
# ============================================================
df = pd.read_excel("student_fitness_lifestyle.xlsx")

# Strip any extra spaces from column names
df.columns = df.columns.str.strip()

# Check columns
print("Columns:", df.columns)

# ============================================================
# Q1: Number of females exercising >6 and <13 hours per week
# ============================================================
q1 = df[(df['Gender'] == 'Female') & 
        (df['Hours_of_Exercise_per_Week'] > 6) & 
        (df['Hours_of_Exercise_per_Week'] < 13)]

print("\nQ1: Number of females practicing 6â€“12 hrs/week:", len(q1))

# Histogram
plt.hist(q1['Hours_of_Exercise_per_Week'], bins=5, color='lightblue', edgecolor='black')
plt.title("Hours of Exercise (Females, 6â€“12 hrs/week)")
plt.xlabel("Hours of Exercise")
plt.ylabel("Number of Females")
plt.show()

# Boxplot
plt.boxplot(q1['Hours_of_Exercise_per_Week'])
plt.title("Boxplot of Hours of Exercise (Females)")
plt.show()

# ============================================================
# Q2: Favorite type of exercise among males
# ============================================================
q2 = df[df['Gender'] == 'Male']['Favorite_Type_of_Exercise'].value_counts()
print("\nQ2: Male Favourite Exercises\n", q2)

# Bar plot
plt.bar(q2.index, q2.values, color='orange', edgecolor='black')
plt.title("Favorite Type of Exercise (Males)")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# ============================================================
# Q3: Min and Max screen time by gender
# ============================================================
min_screen = df['Daily_Screen_Time'].min()
max_screen = df['Daily_Screen_Time'].max()

q3 = df[(df['Daily_Screen_Time'] == min_screen) | 
        (df['Daily_Screen_Time'] == max_screen)]

q3_count = q3.groupby(['Gender', 'Daily_Screen_Time']).size()
print("\nQ3: Min and Max Screen Time by Gender\n", q3_count)

# Prepare labels for x-axis
q3_df = q3_count.reset_index(name='Count')
labels = []
for i in range(len(q3_df)):
    labels.append(q3_df['Gender'][i] + " - " + str(q3_df['Daily_Screen_Time'][i]) + " hrs")

# Bar plot
plt.bar(range(len(q3_df)), q3_df['Count'], color='green', edgecolor='black')
plt.xticks(range(len(q3_df)), labels, rotation=45)
plt.title("Min and Max Screen Time by Gender")
plt.ylabel("Number of Students")
plt.show()

# ============================================================
# Q4: Average and deviation of weekly pocket money spent on snacks
# ============================================================
avg_snacks = df['Weekly_Snacks_Expenditure'].mean()
std_snacks = df['Weekly_Snacks_Expenditure'].std()

print("\nQ4: Average weekly snack expense:", avg_snacks)
print("Q4: Standard deviation:", std_snacks)

# Histogram with average line
plt.hist(df['Weekly_Snacks_Expenditure'], bins=10, color='lightgreen', edgecolor='black')
plt.axvline(avg_snacks, color='red', linestyle='--')
plt.title("Weekly Snack Money Distribution")
plt.xlabel("Money Spent")
plt.ylabel("Number of Students")
plt.show()

# Boxplot
plt.boxplot(df['Weekly_Snacks_Expenditure'])
plt.title("Boxplot of Weekly Snack Money")
plt.ylabel("Money Spent")
plt.show()

# ============================================================
# Q5: Males and females wearing glasses
# ============================================================
q5 = df.groupby(['Gender', 'Wears_Glasses']).size()
print("\nQ5: Students wearing glasses by gender\n", q5)

# Convert to DataFrame for plotting
q5_df = q5.reset_index(name='Count')

# Prepare labels
labels = []
for i in range(len(q5_df)):
    labels.append(q5_df['Gender'][i] + " - " + q5_df['Wears_Glasses'][i])

# Bar plot
plt.bar(range(len(q5_df)), q5_df['Count'], color='purple', edgecolor='black')
plt.xticks(range(len(q5_df)), labels, rotation=45)
plt.title("Students Wearing Glasses by Gender")
plt.ylabel("Number of Students")
plt.show()
```
